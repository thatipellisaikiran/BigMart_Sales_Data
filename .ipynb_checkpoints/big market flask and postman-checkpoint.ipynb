```python
#separating the cat and numerical column
cat,num=[],[]
for keys,values in (df.dtypes).items():
    if values=='O':
        cat.append(keys)
    else:
        num.append(keys)
        
cat,num        
```

```python
cat
```

```python
df.head()
```

```python
from scipy import stats
z_scores = np.abs(stats.zscore(df[num]))
z_score_threshold = 3
# Find the indices of outliers based on the threshold
outlier_indices = np.where(z_scores > z_score_threshold)[0]
# Remove the outliers from the training data
df = df.drop(df.index[outlier_indices])
```

```python
df.shape
```

```python
for categorical in cat:
    from sklearn.preprocessing import LabelEncoder # Import the package
    le=LabelEncoder() # save the package
    df[categorical]=le.fit_transform(df[categorical]) # Apply fit transform
```

# splitting the data


```python
X=df.drop(['Item_Outlet_Sales','Outlet_Establishment_Year'],axis=1)
y=df['Item_Outlet_Sales']
```

```python
#all code into one part
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,    # input data,
                                               y,      #output data
                                               random_state=1234,  #it selects the random samples
                                               test_size=0.30) #test data splitting from original data 30%==03===30%


####################################### step-5 model development  ########################################
#we need to pass the x-tarind and the y train
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(X_train,y_train)

############################################ metrics ##################################
y_pred_LR=LR.predict(X_test)
from sklearn.metrics import r2_score,mean_squared_error
R2_LR=r2_score(y_test,y_pred_LR)
MSE_LR=mean_squared_error(y_test,y_pred_LR)
RMSE_LR=np.sqrt(MSE_LR)
print("rsquare LR:",R2_LR)
print("mse LR:", MSE_LR)
print("rmse LR:",RMSE_LR)
```

```python
 X=df.drop(['Item_Outlet_Sales','Outlet_Establishment_Year'],axis=1)
y=df['Item_Outlet_Sales']
################################
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,# input data
                                               y,#output data random_state=1234,#it selects the random samples
                                               test_size=0.30)#test data splitting from original data 30%==03===30%

#########################################
from sklearn.ensemble import GradientBoostingRegressor
GR=GradientBoostingRegressor(n_estimators=100, learning_rate=0.1,max_depth=3)
GR.fit(X_train,y_train)
y_pred_GR=GR.predict(X_test)
##################################################
from sklearn.metrics import r2_score,mean_squared_error
R2_GR=r2_score(y_test,y_pred_GR)
MSE_GR=mean_squared_error(y_test,y_pred_GR)
RMSE_GR=np.sqrt(MSE_GR)
print("rsquare_GR:",R2_GR)
print("mse_GR:", MSE_GR)
print("rmse_GR:",RMSE_GR)
```

```python
X=df.drop(['Item_Outlet_Sales','Outlet_Establishment_Year'],axis=1)
y=df['Item_Outlet_Sales']
################################
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,# input data
                                               y,#output data random_state=1234,#it selects the random samples
                                               test_size=0.20)#test data splitting from original data 30%==03===30%

#########################################
from catboost import CatBoostRegressor
# Load and preprocess your data
# Initialize and train the CatBoost Regressor
CB= CatBoostRegressor(iterations=100, learning_rate=0.1, depth=6,loss_function='RMSE')
CB.fit(X_train,y_train)
# Make predictions
y_pred_CB= CB.predict(X_test)
#######################################
R2_CB=r2_score(y_test,y_pred_CB)
MSE_CB=mean_squared_error(y_test,y_pred_CB)
RMSE=np.sqrt(MSE_CB)
print("rsquare CB:",R2_CB)
print("mse CB:", MSE_CB)
print("rmse CB:",RMSE)
```
